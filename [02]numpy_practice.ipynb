### 행렬곱 연산 비교
행렬의 **고속 연산** 가능 : python 리스트와 numpy의 연산 속도를 비교해보자.

```python
# 1
a = np.random.normal(0, 0.1, [500, 500]) # 500 x 500 행렬
b = np.random.normal(0, 0.1, [500, 500])

start_time = time.time() 
c = np.matmul(a, b)
print(time.time() - start_time) 
```

```python
# 2
def matmul_py(A, B):
    m = len(A)
    n = len(A[0])
    p = len(B[0])
    C = [[0]*p for i in range(m)]      

    for i in range(0,m):
        for j in range(0,p):
            for k in range(0,n):
                C[i][j] += A[i][k] * B[k][j] 
    return C


A = a.tolist()
B = b.tolist()
start_time = time.time() 
matmul_py(A,B)
print(time.time() - start_time)
```

정방행렬이 아닌 경우 **의사역행렬**을 적용한다.<br>
 image.png

```python
# 3
A = np.array([[1, 2], [2, 4], [3, 6.4]]) 
pinv = np.matmul(lin.inv(np.matmul(A.T, A)), A.T)  
print(A) # 3 x 2
print(pinv) # 2 x 34
```

편의성을 높이기 위해 **`pseudo inverse`** 함수 정의 (정방행렬이 아닌 경우)

```python
# 4
def pinv(A) : # pseudo inverse
    return np.matmul(lin.inv(np.matmul(A.T, A)), A.T)

A = np.array([[1,2], [2,4], [3,6.4]]) 
print(pinv(A))
```

### 주가 지수 예측하기

주가 지수는 이자율에 비례, 실업률에 반비례하다.<br><br>
이자율과 실업률 **2개의 차원**을 가지고 있다.
1. 주가 지수를 예측하는 모델 만들기
2. 실업률과 이자율 변동이 주가 지수에 미치는 영향

1) 이자율과 실업률에 대한 주가 예측 모델은 무엇인가?

2) 실업률이 6.2에서 6.1로 0.1 떨어졌을 때, 이자율이 5.1에서 5.2로 0.1 올라갔을 때 주가에는 어떤 영향이 있는가?

```python
# 1 
Interest_Rate =  np.array([[2.75,2.5,2.5,2.5,2.5,2.5,2.5,2.25,2.25,2.25,2,2,2,1.75,1.75,1.75,1.75,1.75,1.75,1.75,1.75,1.75,1.75,1.75]]).T # 행렬 연산을 위해 두개의 데이터 모두 열벡터로 
Unemployment_Rate = np.array([[5.3,5.3,5.3,5.3,5.4,5.6,5.5,5.5,5.5,5.6,5.7,5.9,6,5.9,5.8,6.1,6.2,6.1,6.1,6.1,5.9,6.2,6.2,6.1]]).T
Stock_Index_Price =  [1464,1394,1357,1293,1256,1254,1234,1195,1159,1167,1130,1075,1047,965,943,958,971,949,884,866,876,822,704,719]

A = np.append(Interest_Rate, Unemployment_Rate, axis = 1) 
A = np.append(A, np.ones((Interest_Rate.size, 1)), axis = 1)
# A 행렬 : 이자율에 대한 열벡터 + 실업률에 대한 열벡터 + 1로 이루어진 열벡터
print(A.shape)

W = np.matmul(pinv(A), Stock_Index_Price)
print(W) # 이자율과 주가 지수는 양의 관계, 실업률과 주가 지수는 음의 관계
```

```python
# 2
x_data = np.array([[5.1, 6.2, 1], [5.1, 6.1, 1], [5.1, 6.2 ,1], [5.2, 6.2, 1] ])
result = np.matmul(x_data, W)
print(result)
```

```python
# 3
print(result[1]-result[0]) # 실업률이 0.1% 떨어지면 주가지수는 25 포인트 상승
print(result[3]-result[2]) # 이자율이 0.1% 올라가면 주가지수는 34 포인트 상승
```
